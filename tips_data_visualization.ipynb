```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.head()
f = plt.figure()
f.savefig('f1.png')
```

```python
tips.info()
f.savefig('f2.png')
```

```python
#histogram
sns.distplot(tips['tip'], kde=False, bins=10)

f.savefig('f2.png')
```

```python
sns.distplot(tips['tip'], kde=False, bins=30);
```

```python
#KDE plot
sns.distplot(tips['tip'], hist=False, bins=10);
```

```python
#scatter plot
sns.relplot(x="total_bill", y="tip", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
#Box Plot
plt.style.use('default')
boxes = plt.boxplot(tips["total_bill"],medianprops={'linewidth':2},patch_artist=True)
for box in boxes['boxes']:
    box.set(color='#4286f4',linewidth=2)
    box.set(facecolor='#e0e0e0')
plt.show()
```

```python
sns.boxplot(x="sex",y="total_bill",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",hue="sex",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",order=["Sun","Sat","Fri","Thur"],data=tips)
```

```python
sns.boxplot(x="sex",y="tip",palette="Blues",data=tips)
```

```python
sns.set()
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", col="time",
            hue="smoker", style="smoker", size="size",
            data=tips)
```
